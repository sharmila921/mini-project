```python
import pandas as pd


#  Load CSV
df = pd.read_csv( r'C:\Users\ADMIN\Downloads\traffic_stops_final5.csv')
```

```python
# Drop the database if it exists
cursor.execute("DROP DATABASE IF EXISTS policelog")
```

```python
#  Create database if not exists
cursor.execute("CREATE DATABASE if not exists policelog")
cursor.execute("USE policelog")
```

```python
#  Create table if not exists
create_table_query = """
CREATE TABLE IF NOT EXISTS traffic_stops (
    stop_date DATE,
    stop_time TIME,
    country_name VARCHAR(50),
    driver_gender CHAR(1),
    driver_age INT,
    driver_race VARCHAR(50),
    violation VARCHAR(100),
    search_type VARCHAR(100),
    stop_outcome VARCHAR(100),
    stop_duration VARCHAR(20),
    drugs_related_stop BOOLEAN,
    vehicle_number VARCHAR(20)
)
"""
cursor.execute(create_table_query)
```

```python

#  Insert rows
insert_query = """
INSERT INTO traffic_stops (
    stop_date,
    stop_time,
    country_name,
    driver_gender,
    driver_age,
    driver_race,
    violation,
    search_type,
    stop_outcome,
    stop_duration,
    drugs_related_stop,
    vehicle_number
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in df.iterrows():
    cursor.execute(insert_query, (
        row['stop_date'],
        row['stop_time'],
        row['country_name'],
        row['driver_gender'],
        row['driver_age'],
        row['driver_race'],
        row['violation'],
        row['search_type'],
        row['stop_outcome'],
        row['stop_duration'],
        row['drugs_related_stop'],
        row['vehicle_number']
    ))

connection.commit()
cursor.close()
connection.close()

print("✅ Data inserted into 'police.traffic_stops' successfully.")
```
